```python
import csv
#urllib is a package that collects several modeules for working with URLs
from urllib.request import urlopen
#Beautiful Soup is a Python library for pulling data out of HTML and XML files
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Downloading content of the web page
html = urlopen("https://en.wikipedia.org/wiki/Comparison_of_programming_languages")
#Creating BeautifulSoup object
soup = BeautifulSoup(html, "html.parser")
#Looking for the table with the class wikitable
table = soup.findAll("table", {"class":"wikitable"})[0]
#Collecting data, "tr" means table row
rows = table.findAll("tr")
```

```python
with open("language.csv", "wt+", newline="") as f:
    writer = csv.writer(f)
    for i in rows:
        row = []
        for cell in i.findAll(["td", "th"]):
            row.append(cell.get_text())
        writer.writerow(row)
```
